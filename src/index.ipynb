### **Welcome to R language**
#### **Variable assignment is of two types:**

```r
x = 876 * 2 * 2 # * Using '='
y <- 100        # * Using '<-' is prefered
```

Historically, before the introduction of `==` for equality testing, `=` was used for that purpose.
Therefore, `<-` became the standard for assignment.
#### **Printing the variables**
- To print out the values associated just type out the variable name
- `print` function is also present

```r
x # Simply prints the output to console
y # Simply prints the output to console
print("Hello, the sum of /nx and y is:")
print(x + y)
```

- `cat` could be use to print custom text

```r
# The '\n' adds a new line
cat("This is another custom message.\nHey!")
print("Hello")
```

##### **Printing custom message using `paste` with `print`:**

```r
my_word <- "example"
print(paste("Here is an", my_word, "word."))
cat("The word is:", my_word, "\n")
```


#### **Creating a vector**
**Vector** is created using **`c`** function
##### **Creating a numeric vector**

```r
numbers <- c(1, 5, 8, 12)
numbers
```


##### **Creating a character vector**

```r
fruits <- c("apple", "banana", "orange")
fruits
```

##### **Creating a logical vector**

```r
status <- c(TRUE, FALSE, TRUE)
status
```


##### **Combining Existing Vectors**

```r
vector1 <- c(10, 20)
vector2 <- c(30, 40)
combined_vector <- c(vector1, vector2)
combined_vector
```


#### **Coercion to a Common Type:**
Vectors converts to highest data types:
1. **logicals** become **numerics**
2. **numerics** become **characters** etc...

```r
mixed_vector <- c(1, TRUE, "hello")
mixed_vector
```


#### **Removing a variable or a group of variable**
There are mainly three ways to remove variables
##### **1. Removing specific objects**

```r
rm(x)
rm(y)
x # * Invalid operation after freeing
y # * Invalid
```


##### **2. Removing multiple objects**

```r
rm(x, y)
x # ! Invalid operation
y # ! Invalid
```


Or using a vector the `list` keyword is a simple data structure to convert it into a list


```r
objects_to_remove <- c("numbers", "fruits",
                       "status", "my_word", "vector1", "vector2",
                       "combined_vector", "mixed_vector")
rm(list = objects_to_remove)
```

##### **3. Remove all objects from the workspace**

```r
rm(list = ls()) # * Closing all memory before finishing
```
