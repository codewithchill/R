### **Some Standard Funtions**
Lets take some sample data
`For Example`:

```r
data <- c(5, 10, 15, 20, 25, 30, 35, 40, 45, 50)
name <- c("Rahul", "Sourav", "Kunal", "Abdul", "Swarnadeep", "Rajat", "Rahul")
movie_lenght <- c("Very Short", "Short", "Medium", "Long",
                  "Very Short", "Very Long", "Short")
print(data)
print(name)
print(movie_lenght)
```

#### **Mean and Standard Deviation**

```r
mean_data <- mean(data)
print(paste("Mean:", mean_data))

std_dev <- sd(data)
print(paste("Standard Deviation:", std_dev))
```

#### **Head function**
Used to print the first `n` elements of a vector

```r
print(head(data, n = 5))
print(head(name, n = 2))
print(head(movie_lenght, n = 3))
```

#### **Factor Function**

```r
ratings <- c(1:10)
print(ratings)
ratings_factor <- factor(ratings)
print(ratings_factor)
name_factor <- factor(name)
print(name_factor)
movie_lenght_factor <- factor(x = movie_lenght,
                              levels = c("Very Short", "Short", "Medium",
                                         "Long", "Very Long"),
                              ordered = TRUE)
print(movie_lenght_factor)
print(levels(movie_lenght_factor))
```

#### **Summary Function**
##### **For Numeric Data**

```r
print(summary(ratings))
print(summary(ratings_factor))
```

##### **For Character Data**

```r
print(summary(name))
print(summary(name_factor))
```

#### **Lenght and Sum Function**

```r
print(length(data))
print(sum(data))
print(length(name))
```

#### **Min and max Function**

```r
print(paste("Max of data:", max(data)))
print(paste("Min of data:", min(data)))
print(paste("Max of name:", max(name)))
print(paste("Min of name:", min(name)))
print(paste("Max of movie_lenght:", max(movie_lenght)))
print(paste("Min of movie_lenght:", min(movie_lenght)))
```

#### **Sum and Sort Function**

```r
dataset <- c(565, 9874, 13215, 21, 6554) # * Temporary dataset
dataset_sum <- sum(dataset)
dataset_sort <- sort(dataset)
cat("Dataset:", dataset, "\n")
print(paste("Sum of dataset:", dataset_sum))
cat("Sorted dataset: ", dataset_sort)
rm(dataset, dataset_sort, dataset_sum) # * Remove when unnecessary
```

### **Some Operations with vector**
#### **Taking vectors with shorthand**

```r
ratings <- c(1:10) # * Creates a vector form 1 to 10
ratings            # * Prints the vector
```

#### **Doing calculation**

```r
ratings <- ratings / 10 #* Dividing all members of ratings by 10
ratings                 #! Prints the values (less used)
print(ratings)          #* Prints it nicely
```

```r
print(ratings[ratings > 0.8]) #* Prints the values greater than 0.8
```

#### **Print the values without assigning**

```r
print(ratings * 100)
rating_01 <- ratings == 0.1 #* Prints the truth values of the
print(rating_01)            #* vector elements mathing the condition
```

#### **Naming a vector using `names` function**

```r
year <- c(2005, 1973, 1977)
print(year)
names(year) <- c("Rahul", "Rajat", "Pallavi")
year <- c(Rahul = 2005,
          Rajat = 1973,
          Pallavi = 1977) # * This the same as above
print(year)
print(names(year))
```

##### **Accessing elements from the vector**

```r
print(year["Pallavi"]) # * Printing the value of pallavi's DOB
print(year[1:2])  # * Prints all the elements between the range(inclusive)
print(year[c("Rajat", "Rahul")])
print(summary(year))   # * Summary of year
```

##### **Operations on elements**

```r
year[1] <- year[1] + 100  # * Add 100 to that specific element
print(year[1])
print(year[3] - year[2])  # * Substract with indexes
print(year["Pallavi"] - year["Rahul"])
```

##### **Checking for a specific condition**

```r
years_gt_2025 <- year > 2025
print(years_gt_2025) # * Rahul will be true because 100 was added
```

##### **Check for the data type of any vector**

```r
print(typeof(year))
print(typeof(name))
print(typeof(data))
```

##### **Checking the condition of a data to be true**

```r
print(is.array(data))
print(is.character(name))
print(is.double(data))
print(is.vector(year)) # * There are many such is. functions
print(is.na(NA))
print(is.na(1 / 0))
print(1 / 0)
```

#### **Doing some inequality operations**

```r
movies_years <- c(1998, 2010, 2016)
print(movies_years > 2014)
print(movies_years == 2015)
print(movies_years != 2015)
```

#### **Special indexing rule**

If we provide a negative index in a `vector` than `R` prints all other values removing that index.

```r
print(movies_years)
print(movies_years[2])
print(movies_years[-2])
```
